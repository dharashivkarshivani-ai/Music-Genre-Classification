```python
# ==========================================
# 🎵 MUSIC CLASSIFICATION SYSTEM - EDA
# ==========================================

# 📦 Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# 🔹 Load the Datasets
df_3s = pd.read_csv("features_3_sec.csv")
df_30s = pd.read_csv("features_30_sec.csv")

# ==========================================
# 1️⃣ BASIC INFO
# ==========================================
print("===== 3 Second Clip Dataset =====")
print(df_3s.info())
print("\n===== 30 Second Song Dataset =====")
print(df_30s.info())

print("\n✅ Missing Values (3 sec):")
print(df_3s.isnull().sum().sum())

print("\n✅ Missing Values (30 sec):")
print(df_30s.isnull().sum().sum())

# ==========================================
# 2️⃣ DATASET SHAPE AND OVERVIEW
# ==========================================
print(f"\n3-sec Dataset Shape: {df_3s.shape}")
print(f"30-sec Dataset Shape: {df_30s.shape}")

print("\nSample Rows (3-sec):")
display(df_3s.head())

# ==========================================
# 3️⃣ GENRE DISTRIBUTION
# ==========================================
plt.figure(figsize=(10,5))
sns.countplot(data=df_3s, x='label', palette='coolwarm', order=df_3s['label'].value_counts().index)
plt.title("🎶 Genre Distribution (3-second Dataset)")
plt.xticks(rotation=45)
plt.xlabel("Music Genre")
plt.ylabel("Count")
plt.show()

# ==========================================
# 4️⃣ SUMMARY STATISTICS
# ==========================================
print("\n📊 Summary Statistics:")
display(df_3s.describe().T)

# ==========================================
# 5️⃣ CORRELATION HEATMAP
# ==========================================
plt.figure(figsize=(12,8))
numeric_features = df_3s.select_dtypes(include=np.number)
sns.heatmap(numeric_features.corr().iloc[:10, :10], annot=True, cmap="coolwarm")
plt.title("🔍 Feature Correlation Heatmap (Top 10 Features)")
plt.show()

# ==========================================
# 6️⃣ BOX PLOTS - EXAMPLE FEATURES
# ==========================================
plt.figure(figsize=(12,6))
sns.boxplot(x='label', y='spectral_centroid_mean', data=df_3s)
plt.title("🎵 Spectral Centroid by Genre")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x='label', y='rms_mean', data=df_3s)
plt.title("🎵 RMS (Energy) by Genre")
plt.xticks(rotation=45)
plt.show()

# ==========================================
# 7️⃣ PCA VISUALIZATION (2D)
# ==========================================
X = df_3s.select_dtypes(include=np.number)
y = df_3s['label']

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['label'] = y

plt.figure(figsize=(10,8))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='label', palette='tab10', s=50, alpha=0.7)
plt.title("🎨 PCA - Genre Clusters (2D Projection)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# ==========================================
# 8️⃣ FEATURE DISTRIBUTION EXAMPLES
# ==========================================
selected_features = ['spectral_centroid_mean', 'tempo', 'rms_mean']
df_3s[selected_features].hist(bins=30, figsize=(10,5))
plt.suptitle("📈 Feature Distributions", fontsize=14)
plt.show()

print("\n✅ EDA Completed Successfully!")
```
